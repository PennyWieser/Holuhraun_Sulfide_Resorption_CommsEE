```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Model 1
df_all=pd.read_csv('Iceland_model1_1.csv')
# Model 2 - Echos Kilauea parameters
cut=3
df=df_all.loc[df_all['pressure']>cut]
df2_all=pd.read_csv('Iceland_model2_6.5.csv')
df2=df2_all.loc[df2_all['pressure']>cut]
```

```python
fig, ((ax0, ax1), (ax3, ax2)) = plt.subplots(2,2, figsize = (12,10))

x='pressure'
y='wS_melt'

#ax0.set_title('Model 1')
ax0.plot(df[x], df[y], '-r', label='Model 1')
ax0.plot(df2[x], df2[y], '-b', label='Model 2')
ax0.set_xlabel('P (MPa)')
ax0.set_ylabel('S melt (ppm)')
ax0.legend()

x='pressure'
y='wH2O_melt'

ax1.plot(df[x], df[y], '-r')
ax1.plot(df2[x], df2[y], '-b')
ax1.set_xlabel('P (MPa)')
ax1.set_ylabel('H$_2$O melt (wt%)')

x='wS_melt'
y='wH2O_melt'
ax2.set_title('Model 2')
ax2.plot(df2['pressure'], 100*df2['wS_melt']/df2['wS_melt'].iloc[0], '-b', label='S')
ax2.plot(df2['pressure'], 100*df2['wH2O_melt']/df2['wH2O_melt'].iloc[0], ':b', label='H2O')
ax2.set_xlabel('P (MPa)')
ax2.set_ylabel('% Degassing')
ax2.legend()
#ax3.set_yscale('log')
#ax3.set_xlim([0, 100])
ax3.legend()
ax3.set_title('Model 1')
ax3.plot(df['pressure'], 100*df['wS_melt']/df['wS_melt'].iloc[0], '-r', label='S')
ax3.plot(df['pressure'], 100*df['wH2O_melt']/df['wH2O_melt'].iloc[0], ':r', label='H2O')
ax3.set_xlabel('P (MPa)')
ax3.set_ylabel('% Degassing')
#ax3.set_yscale('log')
#ax3.set_xlim([0, 100])
ax3.legend()
fig.savefig('SulfurX_Degassing.png', dpi=200)
```
